```python
from sqlalchemy import create_engine, Column, Integer, String, Float, Date, inspect
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import Session
Base = declarative_base()
import pandas as pd
from warnings import filterwarnings
import pymysql
filterwarnings('ignore', category=pymysql.Warning)
import os
from dateutil import parser
import datetime
import time
import dateutil
```

```python
data_file = os.path.join("Resources", "clean_hawaii_measurements.csv")
data_file2 = os.path.join("Resources", "clean_hawaii_stations.csv")
measurements_df = pd.read_csv(data_file)
stations_df = pd.read_csv(data_file2)
```

```python
engine = create_engine('sqlite:///new_database.sqlite')
```

```python
class Station(Base):
    __tablename__ = 'stations'
    id = Column(Integer, primary_key=True)
    station = Column(String(255))
    name = Column(String(255))
    latitude = Column(Float)
    longitude = Column(Float)
    elevation = Column(Integer)

class Measurement(Base):
    __tablename__ = 'measurements'
    id = Column(Integer, primary_key=True)
    station = Column(String(255))
    date = Column(Date)
    prcp = Column(Float)
    tobs = Column(Integer)
```

```python
Base.metadata.create_all(engine)
```

```python
from sqlalchemy.orm import Session
session = Session(bind=engine)
```

```python
measurement_list = list()
for index, row in measurements_df.iterrows():
    measurement = Measurement(station=row['station'], date=parser.parse(row['date']).date(),
                                prcp=row['prcp'], tobs=row['tobs'])
    measurement_list.append(measurement)
```

```python
measurement_list
```

```python
station_list = list()
for index, row in stations_df.iterrows():
    station = Station(station=row['station'], name=row['name'], latitude=row['latitude'],
                                longitude=row['longitude'], elevation=row['elevation'])
    station_list.append(station)
```

```python
session.bulk_save_objects(measurement_list)
session.bulk_save_objects(station_list)
```

```python
session.commit()
```

```python
measurements = session.query(Measurement)
```

```python
for measure in measurements:
    print(measure.id)
    print(measure.station)
```
